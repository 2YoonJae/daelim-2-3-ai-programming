반복문과 조건문 함께 사용

```python
scores =  [42, 60, 77, 89, 20]
for i in scores:
    if i > 60:
        print(f'{i}점은 합격입니다.')
    else :
        print(f'{i}점은 불합격입니다.')
        
    # 60점은 합격입니다.
    # 42점은 불합격입니다.
```

함수

```python
def greet(name):
    print(f'{name}님, 안녕하세요')

greet("대림대")
```

```python
def add_numbers(a, b):
    return a + b

print(add_numbers(1, 2))
```

```python
sum = 0
numbers = [1, 2, 3, 4, 5]
for i in numbers:
    sum  += i
print("합계 : ", sum)
# 합계 : i를 다 더한 값
```

```python
count = 0
scores = [65, 82, 91, 100, 58, 77]

# 70점 이상은 합격
# 출력 : 합격자는 x명 입니다.
for a in scores :
    if a >= 70:
        count += 1
print(f'합격자는 {count}명 입니다.')
```
