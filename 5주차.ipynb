# 복습

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data = {
    "study_hours" : [1, 2, 3, 3, 4, 4, 5, 6, 7, 8],
    "attendance" : [50, 55, 60, 80, 60, 75, 75, 80, 90, 95],
    "pass_fail" : [0, 0, 0, 1, 0, 1, 1, 1, 1, 1] 
}
df = pd.DataFrame(data)

# 입력값과 정답 분리
X = df[["study_hours", "attendance"]]
y = df["pass_fail"]

model = DecisionTreeClassifier(random_state=42)
model.fit(X, y)

print("데이터 준비 완료")
print(f"학생 수 :{len(df)}명") #학생 수 : %%명
```

# 모델 정확도

- 학습에 사용한 데이터로 정확도 확인

```python
accuracy = model.score(X, y)

print("모델 정확도:", accuracy) # 0~1 표시됨 -> 0~100%
print(f"모델 정확도 퍼센트: {accuracy*100}%")
```

```python
# 모델을 어떤 예측을 했는지 확인
predictions = model.predict(X)

# Columns - 공부시간, 출석률, 실제, 예측, 결과
# zip() => zip([1, 2, 3], [4, 5, 6]) => (1, 4), (2, 5), (3,6) -서로가 한 쌍
# 실제값과 예측값을 한쌍으로 묶어서...
# for a, p in zip(y, predictions):
#   if (a==p):
#       print("0")
#   else
#       print("~~")

result = pd.DataFrame({
    "공부시간": X["study_hours"].values,
    "출석률": X["attendance"].values,
    "실제": y.values,
    "예측": predictions,
    "결과": ["O" if a == p else "X" for a , p in zip(y, predictions)]
})

result
```

## 문제

- 사과, 바나나, 포도는 각 1000원, 2000원, 3000원 이다.
    - 딕셔너리 로그 출력
- 사과의 가격을 10% 할인하고 바나나는 13% 인상했을 때,
    - Python 코드로 작성
- 사과, 바나나, 포도의 가격표를 표시하세요. 
    - 로그 출력 (DataFrame 사용)

```python
fruits = {
    "사과" : 1000,
    "바나나" : 2000,
    "포도" : 3000
}
fruits_dic = {x: f"{y}원" for x, y in fruits.items()} # 딕셔너리 컴프리헨션
print(fruits_dic)   #딕셔너리 로그 출력(인상 전)

fruits["사과"] = int(fruits["사과"] * 0.9) #10% 할인
fruits["바나나"] = int(fruits["바나나"] * 1.13) #13% 인상

fruits_dic = {x: f"{y}원" for x, y in fruits.items()}
print(fruits_dic)   #딕셔너리 로그 출력(인상 후)

fruits_df = pd.DataFrame({
    "과일" : fruits.keys(),
    "가격" : [f"{x}원" for x in fruits.values()] #리스트 컴프리헨션
})

fruits_df #로그 출력 (DataFrame)
```

# 날씨 예측

```python
#데이터 준비

weather_data = {
    "clouds" : [10, 20, 55, 50, 60, 70, 80, 90, 95],
    "humidity" : [30, 35, 60, 60, 65, 70, 80, 85, 90],
    "umbrella" : [0, 0, 0, 0, 1, 1, 1, 1, 1]
}

weather_df = pd.DataFrame(weather_data)
weather_df
```

```python
# 입력값과 정답 분리

X_weather = weather_df[["clouds", "humidity"]]
y_weather = weather_df["umbrella"]

print("입력값과 정답 분리 완료!")
```

```python
weather_model = DecisionTreeClassifier(random_state=42)
weather_model.fit(X_weather, y_weather)
print("모델 학습 완료")
```

```python
test_weather = pd.DataFrame([
    [20, 35], # 구름 적음, 습도 낮음
    [55, 65], # 구름 보통, 습도 보통
    [85, 88], # 구름 많음, 습도 높음
], columns=["clouds", "humidity"])

predictions = weather_model.predict(test_weather)
test_weather
```

## 문제 
아래 로그 개별 출력 
- 구름 ?%, 습도 ?% -> 우산 필요|불필요
    - 예시) 구름 20%, 습도  35% -> 우산 불필요

```python
for i in range(len(test_weather)) :
    if predictions[i] == 1 :
        print(f"구름 {test_weather.iloc[i, 0]}%, 습도{test_weather.iloc[i, 1]}% -> 우산 필요")
    else :
        print(f"구름 {test_weather.iloc[i, 0]}%, 습도{test_weather.iloc[i, 1]}% -> 우산 불필요")
```

```python
# 모델 정확도 구하기

acc = weather_model.score(X_weather, y_weather)
print(f"모델 정확도: {acc*100}%")
```

```python
# 모델 정확도 낮추기
# 애애한 데이터를 포함한 날씨 데이터

messy_data = {
    "clouds" : [10, 20, 55, 55, 60, 70, 80, 90],
    "humidity" : [30, 35, 60, 60, 65, 70, 80, 85],
    "umbrella" : [0, 0, 0, 1, 1, 1, 1, 1],
}

messy_df = pd.DataFrame(messy_data)

X_messy = messy_df[["clouds", "humidity"]]
y_messy = messy_df["umbrella"]

messy_model = DecisionTreeClassifier(random_state=42)
messy_model.fit(X_messy, y_messy)

messy_acc = messy_model.score(X_messy, y_messy)

print(f"애매한 데이터 모델 정확도: {messy_acc * 100}%")

print()

messy_pred = messy_model.predict(X_messy)
result = pd.DataFrame({
    "구름": X_messy["clouds"].values,
    "습도": X_messy["humidity"].values,
    "실제": y_messy.values,
    "예측": messy_model.predict(X_messy),
    "결과": ["o" if a == p else "x" for a, p in zip(y_messy, messy_pred)]})

result
```
